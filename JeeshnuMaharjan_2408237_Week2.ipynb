```python
# Problem 1
import pandas as pd

# loading and reading the csv file into bank_df
bank_df = pd.read_csv('/content/drive/MyDrive/year2/Datasets/bank .csv')

# displays short summery of the data table
print(bank_df.info())

# a. identifies and selects columns with dtype='object'
object_columns = bank_df.select_dtypes(include=['object']).columns
print("\nColumns with dtype='object':\n", object_columns)

# b. finds unique values of those columns
for col in object_columns:
    print(f"\nUnique values in column '{col}':", bank_df[col].unique())

# c. checks number of null value in all cols
null_values = bank_df.isnull().sum()
print("\nNumber of null values in each column:\n", null_values)

# drops columns with dtype='object' and store the result
banknumericdata = bank_df.drop(columns=object_columns)

# saves the DataFrame with numeric data to a CSV file
banknumericdata.to_csv('banknumericdata .csv', index=False)
print("\nNumeric data saved to 'banknumericdata.csv'.")

# reads the saved CSV and generate summary statistics
banknumericdata = pd.read_csv('banknumericdata .csv')
summary_stats = banknumericdata.describe()
print("\nSummary statistics of numeric data:\n", summary_stats)
```

```python
import pandas as pd

# reading and loading csv file
medical_df = pd.read_csv('/content/drive/MyDrive/year2/Datasets/medical_students_dataset.csv')

# finding number of null value per col
print("Initial Missing Values Per Column:")
print(medical_df.isnull().sum())

# filling the null value using mode or median
for column in medical_df.columns:
    if medical_df[column].isnull().sum() > 0:
        if medical_df[column].dtype == 'object':
            medical_df[column] = medical_df[column].fillna(medical_df[column].mode()[0])
        else:
            medical_df[column] = medical_df[column].fillna(medical_df[column].median())

print("\nMissing Values After Handling:")
print(medical_df.isnull().sum())

# deleting duplicate rows
print(f"\nNumber of duplicate rows: {medical_df.duplicated().sum()}")
medical_df = medical_df.drop_duplicates()
print(f"Number of duplicate rows after removal: {medical_df.duplicated().sum()}")

print("\nCleaned dataset saved successfully.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# PROBLEM 1
# reading the Titanic dataset
titanic_df = pd.read_csv('/content/drive/MyDrive/year2/Datasets/Titanic-Dataset.csv')

# getting the subset the DataFrame with the specified columns
df_subset = titanic_df[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]

# Filter rows where Pclass is equal to 1 (first-class passengers)
df_first_class = df_subset[df_subset['Pclass'] == 1]

# getting mean, median, max, and min of the 'Fare' column
mean_fare = df_first_class['Fare'].mean()
median_fare = df_first_class['Fare'].median()
max_fare = df_first_class['Fare'].max()
min_fare = df_first_class['Fare'].min()

# displaying the results
print(f"Mean Fare: {mean_fare}")
print(f"Median Fare: {median_fare}")
print(f"Max Fare: {max_fare}")
print(f"Min Fare: {min_fare}")



# PROBLEM 2
# checks sum of null values in the 'Age' column
null_age_count = df_subset['Age'].isnull().sum()

# displaying the value
print(f"Number of null values in 'Age' column: {null_age_count}")

# dropping null value
df_subset_cleaned = df_subset.dropna(subset=['Age'])



# PROBLEM 3
# Step 1: Using one-hot encoding for the 'Embarked' column
embarked_dummies = pd.get_dummies(titanic_df['Embarked'], prefix='Embarked')

# Step 2: adding these new columns to the original DataFrame
titanic_df = pd.concat([titanic_df, embarked_dummies], axis=1)

# Step 3: deleting the original 'Embarked' column
titanic_df.drop('Embarked', axis=1, inplace=True)

# converting into 1/0 from True/False
titanic_df['Embarked_C'] = titanic_df['Embarked_C'].astype(int)
titanic_df['Embarked_Q'] = titanic_df['Embarked_Q'].astype(int)
titanic_df['Embarked_S'] = titanic_df['Embarked_S'].astype(int)

# Step 4: displaying the first few rows to verify the changes
print(titanic_df.head())



# PROBLEM 4
# calculating the mean survival rate by gender
mean_survival_by_gender = titanic_df.groupby('Sex')['Survived'].mean()

# creating a bar plot (using pandas' built-in plot)
print()
mean_survival_by_gender.plot(kind='bar', color=['blue', 'orange'], figsize=(8, 6))

# adding title and labels
plt.title('Mean Survival Rates by Gender')
plt.xlabel('Gender')
plt.ylabel('Mean Survival Rate')

# displaying the plot
plt.show()


# PROBLEM 5
passengers_by_port = titanic_df[['Embarked_C', 'Embarked_S', 'Embarked_Q']].sum()

# making a graph using panda
passengers_by_port.plot(kind='bar', color=['blue', 'green', 'orange'], figsize=(8, 6), title="Number of Passengers by Port of Embarkation")
```
