```python
# 3.1 Warm Up Exercises:
# SORTING

import pandas as pd

# loading and reading the dataset
titanic_df = pd.read_csv('/content/drive/MyDrive/year2/Datasets/Titanic-Dataset.csv')

# creating a dataFrame for Fare and display first 5 rows
fare = titanic_df[['Fare']]
print(fare.head())

# creating a dataFrame for Pclass and Age and display first 5 rows
class_age = titanic_df[['Pclass', 'Age']]
print(class_age.head())

# creating a dataFrame for Survived and Sex and display first 5 rows
survived_gender = titanic_df[['Survived', 'Sex']]
print(survived_gender.head())
```

```python
# SUBSETTING

# finding rows with fare greater than 100
fare_gt_100 = titanic_df[titanic_df['Fare'] > 100]
print(fare_gt_100)

# finding rows where Pclass is 1
first_class = titanic_df[titanic_df['Pclass'] == 1]
print(first_class)

# finding rows where Age is less than 18 and Sex is female
female_under_18 = titanic_df[(titanic_df['Age'] < 18) & (titanic_df['Sex'] == 'female')]
print(female_under_18)

# finding rows where Embarked is either 'C' or 'S'
embarked_c_or_s = titanic_df[titanic_df['Embarked'].isin(['C', 'S'])]
print(embarked_c_or_s)

# finding rows where Pclass is either 1 or 2
first_second_class = titanic_df[titanic_df['Pclass'].isin([1, 2])]
print(first_second_class)
```

```python
# 3.2 Exploratory Data Analysis Practice Exercise:

# filling missing values in Age with median
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].median())

# adds a new column 'Fare per year'
titanic_df['Fare per year'] = titanic_df['Fare'] / titanic_df['Age']

# subsets rows where Fare per year > 5 and sort in descending order
high_fare_age = titanic_df[titanic_df['Fare per year'] > 5]
high_fare_age_srt = high_fare_age.sort_values(by='Fare per year', ascending=False)

# displays Name and Fare per year for these rows
result = high_fare_age_srt[['Name', 'Fare per year']]
print(result)

# adds a column 'Fare per class'
titanic_df['Fare per class'] = titanic_df['Fare'] / titanic_df['Pclass']

# subsets for adult males and sort in descending order by Fare per class
adult_males = titanic_df[(titanic_df['Age'] >= 18) & (titanic_df['Sex'] == 'male')]
adult_males_srt = adult_males.sort_values(by='Fare per class', ascending=False)

# displays 'Name', 'Age', and 'Fare per class' for these rows
result_adult_males = adult_males_srt[['Name', 'Age', 'Fare per class']]
print(result_adult_males)
```

```python
# 3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:

# calculates total 'Fare' revenue
total_fare = titanic_df['Fare'].sum()

# calculates 'Fare' revenue for each class
fare_first_class = titanic_df[titanic_df['Pclass'] == 1]['Fare'].sum()
fare_second_class = titanic_df[titanic_df['Pclass'] == 2]['Fare'].sum()
fare_third_class = titanic_df[titanic_df['Pclass'] == 3]['Fare'].sum()

# calculates fare proportions
fare_proportions = [
    fare_first_class / total_fare * 100,
    fare_second_class / total_fare * 100,
    fare_third_class / total_fare * 100
]
print("percentage of fare revenue by class:", fare_proportions)

# functions to categorize passengers by age group
def categorize_age(age):
    if age < 18:
        return 'child'
    elif age < 65:
        return 'adult'
    else:
        return 'senior'


# adds a new column 'Age Group'
titanic_df['Age Group'] = titanic_df['Age'].apply(categorize_age)

# calculates total passengers and count by age group
total_passengers = len(titanic_df)
age_group_counts = titanic_df['Age Group'].value_counts()

# calculates proportions of passengers in each age group
age_group_proportions = (age_group_counts / total_passengers) * 100
print("percentage of passengers by age group:")
print(age_group_proportions)
```
